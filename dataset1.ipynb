```python
#ici on importe le fichier de la quatrieme visualisatoin
file_path=r"C:/Users/a/Desktop/projet bokeh/annuaire-des-guides.csv"
df_guide=pd.read_csv(file_path, encoding='UTF-16',delimiter='\t', on_bad_lines='skip')
```

```python
df_guide=df_guide.drop_duplicates(subset=['Nom','Prénom','Ville','Langue de travail'])
```

```python
df_guide[80:95]
```

```python
#on supprime les lignes qui ont Nan dans la colonne langue de travail
df_guide = df_guide.dropna(subset=['Langue de travail'])
```

```python
df_guide[80:95]
```

```python
#ici il y avait dans le tableau des guides qui avait la colonne vide alors on l a rempli en repetant le Nom
df_guide['Prénom']=df_guide['Prénom'].fillna(df_guide['Nom'])
```

```python
df_guide[80:95]
```

```python
#pour savoir les langues qu on a
unique_languages=df_guide['Langue de travail'].unique()
```

```python
unique_languages
```

```python
pd.set_option('display.max_rows', None)
langue_counts=df_guide['Langue de travail'].value_counts()
langue_counts
```

```python
#ici on a remarque que plusieurs langue se repetent mais elles sont ecrites d une maniere differentes donc on a essayer de les regrouper
df_guide['Langue de travail']=df_guide['Langue de travail'].replace({'ANG-FR': 'FR-ANG', 'RUSS': 'RUSSE'})
```

```python
langue_counts=df_guide['Langue de travail'].value_counts()
langue_counts
```

```python
df_guide['Langue de travail'] = df_guide['Langue de travail'].replace({'HOLL': 'NEER', 'RUS': 'RUSSE','ESP-ANG':'ANG-ESP','FR-RUSS':'FR-RUSSE','ANG-FR-ALL':'FR-ANG-ALL','Fr - Ang':'FR-ANG','ITA-FR':'FR-ITA'})
```

```python
# il y a des langues qui ne se repetent qu une seule fois alors on a supprimer qui se repetent moins de 2 fois
langue_counts=df_guide['Langue de travail'].value_counts()
langues_a_garder=langue_counts[langue_counts>2].index
df_guide=df_guide[df_guide['Langue de travail'].isin(langues_a_garder)]
```

```python
df_guide['Langue de travail']=df_guide['Langue de travail'].replace({'ITA-FR':'FR-ITA'})
```

```python
langue_counts=df_guide['Langue de travail'].value_counts()
langue_counts
```

```python
#ici on a exporter notre dataframe sous forme d un fichier excel
df_guide.to_excel('C:/Users/a/Documents/fichierg.xlsx',index=False)
```

```python
#pour creer la derniere visualisation il fallait modifier la forme de notre dataframe alors on importer de nouveau le fichier qu on vient d exporter

df_guide_f=pd.read_excel('C:/Users/a/Desktop/fichierg.xlsx')
#on regroupe par ville
df_2=df_guide_f.groupby('Ville').size().reset_index(name='Nombre de guides')
#on ajouter les langues sous forme de colonnes
langue_counts=df_guide_f.pivot_table(index='Ville',columns='Langue de travail',aggfunc='size',fill_value=0)
#on fusionne dans une nouvelle dataframe
df_final=df_2.merge(langue_counts,on='Ville',how='left')

```
